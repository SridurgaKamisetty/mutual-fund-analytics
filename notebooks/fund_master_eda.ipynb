```python
import pandas
nav_history=pandas.read_csv(path + "02_nav_history.csv")
fund_master_codes=set(fund_master['amfi_code'].unique())
nav_history_codes=set(nav_history['amfi_code'].unique())
```

```python
missing_codes=fund_master_codes - nav_history_codes
```

```python
print("Total codes in fund master:",len(fund_master_codes))
print("Total codes in nav history:",len(nav_history_codes))
print("Total missing codes:",len(missing_codes))
print("missing codes list:", missing_codes)
```

```python
#with merge

merged = fund_master.merge(nav_history[['amfi_code']].drop_duplicates(), 
                             on='amfi_code', how='left', indicator=True)

missing_count = (merged['_merge'] == 'left_only').sum()

print(f"Missing codes count: {missing_count}")
```
